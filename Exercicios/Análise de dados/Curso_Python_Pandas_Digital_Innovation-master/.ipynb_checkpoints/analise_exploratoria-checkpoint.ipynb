```python
#Step one: Importing the libraries
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use("seaborn")

#Step two: Uploading the file
file_name = "datasets/AdventureWorks.xlsx"

#Step three: Creating the data frame
df = pd.read_excel(file_name, engine="openpyxl")

df["Custo"] = df["Custo Unitário"].mul(df["Quantidade"])
custo_total = round(df["Custo"].sum(), 2)

df["Lucro"] = df["Valor Venda"] - df["Custo"]
lucro_total = round(df["Lucro"].sum(),2)

print("Question four: ")
print("What was the total profit?")
print("Answer: {}" .format(lucro_total))
print("\n")

print("Question five: ")
print("Calculating the shipping date")
print("Calculando o tempo de envio")
#df["Tempo_Envio"] = df["Data Envio"] - df["Data Venda"] #Presents with description (String)
df["Tempo_Envio"] = (df["Data Envio"] - df["Data Venda"]).dt.days #Presents without description (Interger)
print("The type of df[Tempo_Envio] is: {}" .format(df["Tempo_Envio"].dtype))
print(df.head(1))
print("\n")

print("Question six: ")
print("Calculating average shipping time by brand")
print("Calculando tempo médio de envio por marca")
print("Answer: {}" .format(df.groupby(df["Marca"])["Tempo_Envio"].mean()))
print("\n")

print("Question seven: ")
print("Adding the number of cells per column with null values")
print("Somando a quantidade de células por coluna com valores nulos/vazios")
df.isnull().sum()
```

```python
#Step one: Importing the libraries
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use("seaborn")

#Step two: Uploading the file
file_name = "datasets/AdventureWorks.xlsx"

#Step three: Creating the data frame
df = pd.read_excel(file_name, engine="openpyxl")

df["Custo"] = df["Custo Unitário"].mul(df["Quantidade"])
df["Lucro"] = df["Valor Venda"] - df["Custo"]

df_2009 = df[df["Data Venda"].dt.year == 2009]#Creating a new data frame
df_2009.groupby(df["Data Venda"].dt.month)["Lucro"].sum().plot(title="Lucro x Mês");
plt.xlabel("Mês")
plt.ylabel("Lucro");
```
